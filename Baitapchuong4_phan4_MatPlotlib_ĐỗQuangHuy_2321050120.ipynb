```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('gas_prices.csv')
data.head()
```

```python
year = list(data['Year'])
gas = list(data['Japan'])
print('Năm: ', year)
print('Giá gas của japan: ', gas)
```

```python
plt.figure(figsize = (12,4))
plt.plot(year,
          gas,
          'c-.d')
plt.title('Biểu đồ giá Gas tại Nhật giai đoạn 1990 - 2008')
plt.ylabel('Giá Gas (USD)')
plt.xlim(1989,2009)
plt.ylim(2.5,6.0)
plt.xticks(year, rotation=45)
plt.grid(axis = 'y',
        c = 'black',
        ls = '-.',
        lw = 0.5)
plt.show()
```

```python
year = list(data['Year'])
aus = list(data['Australia'])
hq = list(data['South Korea'])
uk = list(data['UK'])
usa = list(data['USA'])
print('Năm: ', year)
print('Giá gas của Australia: ', aus)
print('Giá gas của South Korea: ', hq)
print('Giá gas của UK: ', uk)
print('Giá gas của USA: ', usa)
```

```python
plt.figure(figsize = (12,4))
plt.plot(year,
         aus,
         'c--',
         label='Australia')
plt.plot(year,
         hq,
         'y--',
         label='South Korea')
plt.plot(year,
         uk,
         'r-^',
         label='UK')
plt.plot(year,
         usa,
         c='black',
         ls='-',
         marker='v',
         label='USA')
plt.title('Gas Prices over time 1990 - 2008')
plt.ylabel('Dollars ($)')
plt.xlim(1989,2009)
plt.ylim(0,8)
plt.xticks(year, rotation=45)
plt.legend(loc=4, ncol=2)
plt.grid(axis = 'x',
        c = 'black',
        ls = '--',
        lw = 0.5)
plt.show()
```

```python
labels=['K60', 'K61', 'K62', 'K63', 'K64','K65']
boys = [200,340, 260, 440, 300, 180]
girls = [30, 60, 90, 160, 180, 220]
total = list (np.array(boys) + np.array(girls) )
sex = ['nam', 'nữ']
sum_boy = sum(boys)
sum_girl = sum(girls)
sumbg = [sum_boy, sum_girl]
tong=np.array([boys[i] + girls[i] for i in range (0, len(boys))])
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,2,1)
plt.title('Biểu đồ tỷ lệ SV nam - nữ khoa CNTT')
plt.pie(sumbg, 
        labels = sex,
        autopct = '%.2f%%',
        pctdistance = 0.7,
        startangle = 90,
        labeldistance = 1.05,
        textprops = {'color':'w','fontweight':'bold'},
        wedgeprops = dict(edgecolor = 'w'),
        rotatelabels =False)
plt.legend(ncol = 1,loc = 4)

plt.subplot(1,2,2)
plt.title('Biểu đồ số lượng SV theo từng khóa')
bar = plt.bar(labels , tong)
bar[3].set_color('orange')
plt.grid(axis = 'y',ls = '--')
plt.show()
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,2,1)
plt.title('Biểu đồ số lượng SV nam, nữ và tổng theo từng khóa')
plt.plot(labels,
         boys,
         'b--o',
         label='boys')
plt.plot(labels,
         girls,
         'k--o',
         label='girls')
plt.plot(labels,
         tong,
         'r-s',
         label='tong')

plt.ylim(0,625)
plt.xticks(labels, rotation=0)
plt.legend(loc=4, ncol=1)
plt.grid(axis = 'x',
         c = 'black',
         ls = '--',
         lw = 0.5)
plt.grid(axis = 'y',
         c = 'black',
         ls = '--',
         lw = 0.5)
plt.subplot(1,2,2)
plt.title('Biểu đồ tỷ lệ SV của từng khóa')
e = [0,0.2,0,0.2,0,0]
plt.pie(tong, 
        labels = labels,
        explode=e,
        autopct = '%.2f%%',
        pctdistance = 0.75,
        startangle = 180,
        labeldistance = 1.05,
        textprops = {'color':'w','fontweight':'bold'},
        wedgeprops = dict(width=0.5, edgecolor = 'w'),
        shadow = True)
plt.legend(ncol = 2,loc = 4)
plt.show()
```

2. Số lượng người từ 30-42 tuổi chiếm khoảng 38.89%
3. Số lượng người từ 42-58 tuổi chiếm khoảng 13.78%
4. Số lượng người còn lại chiếm khoảng 0.66%
5. Điều này phản ánh cơ cấu dân số trong dữ liệu nghiêng mạnh về người trẻ và trung niên, trong khi nhóm cao tuổi rất ít.
6. Phân bố độ tuổi có dạng lệch phải (skewed right), tập trung nhiều ở nhóm trẻ và trung niên, thưa dần về phía nhóm cao tuổi.

```python
import warnings
import numpy as np
warnings.filterwarnings('ignore')
data['Index'] = data['Index'].replace({0: 'Extremely Weak',1: 'Weak',2: 'Normal',3: 'Overweight',4: 'Obesity',5: 'Extreme obesity'})
order = ['Extreme obesity', 'Obesity', 'Overweight', 'Normal', 'Weak', 'Extremely Weak']
data['Index'] = pd.Categorical(data['Index'], categories = order, ordered = True)
status_count = data.groupby(['Index', 'Gender']).size().unstack(fill_value = 0)
plt.figure(figsize = (12, 4.8))
plt.title('BIỂU ĐỒ THỐNG KÊ TÌNH TRẠNG CƠ THỂ THEO GIỚI TÍNH', fontdict = {'fontname': 'Arial', 'fontweight': 'bold', 'fontsize': 14})
labels = status_count.index
x = np.arange(len(labels))
w = 0.4 
plt.bar(x - w/2, status_count['Male'], width = w, color = 'orange', label = 'Male')
plt.bar(x + w/2, status_count['Female'], width = w, color = 'green', label = 'Female')
plt.ylabel('Số lượng', fontsize=12)
plt.xticks(x, labels)
plt.grid(axis = 'y', linestyle = '--')
plt.legend()
plt.show()
```

## Nhận xét: Từ biểu đồ, ta có thể nhận thấy rằng:
1. nam có số lượng cao hơn trong các tình trạng béo phì
2. nữ có thể chiếm ưu thế hơn trong các tình trạng gầy

```python
height = data.loc[:,'Height']
weight = data.loc[:,'Weight']
plt.figure(figsize = (12, 5.4))
plt.title('BIỂU ĐỒ THỂ HIỆN MỖI QUAN HỆ GIỮA CHIỀU CAO VÀ CÂN NẶNG TRONG TẬP DỮ LIỆU',fontdict={'fontname': 'Arial', 'fontweight': 'bold', 'fontsize': 14})
plt.scatter(height,weight,marker = 'o',s = 120,alpha = 0.7,color = 'green')
plt.grid(ls='--')
plt.xlabel('Chiều cao(cm)')
plt.ylabel('Cân nặng(kg)')
plt.show()
```

1. Phân bố dữ liệu:
    
        Chiều cao dao động trong khoảng 140cm đến 200cm.
        Cân nặng dao động trong khoảng 50kg đến 160kg.
        Các điểm dữ liệu phân bố khá dày và rải đều, không tập trung vào một vùng cụ thể.

2. Mối quan hệ giữa chiều cao và cân nặng:

        Không có xu hướng rõ rệt tuyến tính (tăng hay giảm đồng thời).
        Cùng một chiều cao, cân nặng có thể thay đổi rất nhiều (ví dụ: người cao 170cm có cân nặng từ khoảng 60kg đến 150kg).

3. Đặc điểm nổi bật:

        Dữ liệu thể hiện sự đa dạng về thể trạng: có người thấp nhưng cân nặng cao, và ngược lại.
        Phân bố khá đồng đều trên toàn bộ khung dữ liệu, cho thấy tập dữ liệu chứa nhiều mẫu với đặc điểm phong phú.
