```python
df_patient = pd.read_csv('Data_Patient.csv',index_col = 0)
df_patient.rename(columns = {'feature_1':'Age',
                             'feature_2':'Gender',
                             'feature_3':'Type',
                             'feature_4':'Blood_pressure',
                             'feature_5':'Cholesterol',
                             'feature_6':'Heartbeat',
                             'feature_7':'Thalassemia',
                             'feature_8':'Result'},inplace = True)
df_patient.head()
```

Yêu cầu 4.2

```python
male = df_patient[df_patient['Gender']=='Male']
female = df_patient[df_patient['Gender']=='Female']
no = df_patient[df_patient['Result']==0]
yes = df_patient[df_patient['Result']==1]
print('đàn ông:\n',male.head())
print('Phụ nữ:\n',female.head())
print('no\n',no.head())
print('yes\n',yes.head())
```

Yêu cầu 4.3

```python
data1 = df_patient[(df_patient['Result'] == 1) & (df_patient['Age'] >= 70)]
data2 = df_patient[(df_patient['Gender'] == 'Female') & (df_patient['Blood_pressure'] >= 170) & (df_patient['Result'] == 0)]
data3 = df_patient[(df_patient['Type'] == 'Typical angina') & (df_patient['Gender'] == 'Male') & (df_patient['Result'] == 1)]
print('Lọc 1:\n',data1)
print('Lọc 2:\n',data2)
print('Lọc 3:\n',data3)
```

Yêu cầu 4.4

```python
print('1. Chỉ số huyết áp:')
print('Thấp nhất:',df_patient['Blood_pressure'].min())
print('Cao nhất:',df_patient['Blood_pressure'].max())
print('Trung bình:',df_patient['Blood_pressure'].mean())
print('Trung vị:',df_patient['Blood_pressure'].median())
print('Độ lệch chuẩn:',df_patient['Blood_pressure'].std())
print('2. Chỉ số nhịp tim:')
print('Thấp nhất:',df_patient['Heartbeat'].min())
print('Cao nhất:',df_patient['Heartbeat'].max())
print('Trung bình:',df_patient['Heartbeat'].mean())
print('Trung vị:',df_patient['Heartbeat'].median())
print('Độ lệch chuẩn:',df_patient['Heartbeat'].std())
```
